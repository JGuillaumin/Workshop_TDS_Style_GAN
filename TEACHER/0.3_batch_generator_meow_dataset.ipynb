```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from gan_utils import BatchGenerator
```

```python
path_dataset = '../data/cat_dataset/'
```

```python
batch_gen = BatchGenerator(path_dataset, batch_size=16, color_mode='rgb',
                           target_size=(128,128), pre_processing="unit")
```

```python
x = batch_gen.next()
print("x : {}  |  {}".format(x.shape, x.dtype))
_ = plt.hist(x.flatten(), normed=True, bins=255)
```

```python
x = batch_gen.next()
fig, axes = plt.subplots(figsize=(10, 10), ncols=2, nrows=2)
axes[0,0].imshow(np.clip(x[0]*255, 0, 255).astype(np.uint8))
axes[0,1].imshow(np.clip(x[1]*255, 0, 255).astype(np.uint8))
axes[1,0].imshow(np.clip(x[2]*255, 0, 255).astype(np.uint8))
axes[1,1].imshow(np.clip(x[3]*255, 0, 255).astype(np.uint8))
```
